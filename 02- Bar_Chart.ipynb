### How to set order

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

#make a line plot
sns.barplot(x ="sex", y= "alone", data= kashti, hue="who", order=["female","male"]) #order is used for variable order
plt.show()
```

### Order Palette

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

#make a line plot
sns.barplot(x ="sex", y= "alone", data= kashti, hue="who", order=["female","male"], color = "red") #color is used for different colors
plt.show()
```

### Remove confidence interval

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

#make a line plot
sns.barplot(x ="sex", y= "alone", data= kashti, hue="who", order=["female","male"], color = "red", ci= None) #ci means confidence interval
plt.show()
```

### Palette

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

#make a line plot
sns.barplot(x ="sex", y= "alone", data= kashti, hue="who", order=["female","male"], color = "red", ci= None,
           palette ="deep") 
plt.show()
```

### Estimator

```python
# import libraries
import seaborn as sns
from numpy import mean
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

#make a line plot
sns.barplot(x ="sex", y= "alone", data= kashti, hue="who", order=["female","male"], color = "red", ci= None,estimator= mean
           ) 
plt.show()
```

### Color saturation

```python
# import libraries
import seaborn as sns
import numpy
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

#make a line plot
sns.barplot(x ="class", y= "fare", data= kashti, hue="sex", estimator= mean, saturation=2)
#saturation is used to light the color as you increase the saturation the graph will get brighter            
plt.show()
```

```python
# import libraries
import seaborn as sns
import numpy
import matplotlib.pyplot as plt

# load dataset
kashti = sns.load_dataset("titanic")
kashti

#make a line plot
sns.barplot(x ="class", y= "fare", data= kashti, hue="sex", estimator= mean, saturation=0.5)
#saturation is used to light the color as you increase the saturation the graph will get brighter            
plt.show()
```

### Different types of graphs

```python
# Importing the required libraries

import seaborn as sns
import matplotlib.pyplot as plt

#read the titanic data
#from seaborn library

kashti = sns.load_dataset("titanic")

sns.barplot(x ="class", y = "fare", data= kashti,
           linewidth = 2.5, facecolor= (0.5,0.5,0.3,0.5),
           errcolor= ".2", edgecolor = ".2")
#err o means full black and 1 means full white
#edge color 0 means full black and 1 means white 
```
