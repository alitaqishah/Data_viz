### Import Libraries
Seaborn automatically install these libraries.
- *numpy*
- *scipy*
- *pandas*
- *matplotlib*

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

#make a line plot
sns.lineplot(x ="sepal_length", y= "sepal_width", data= phool)
plt.show()
```

### Adding title

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

#make a line plot
sns.lineplot(x ="sepal_length", y= "sepal_width", data= phool)
plt.title("Phoolo ka plot")
plt.show()
```

### Adding limits

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

#make a line plot
sns.lineplot(x ="sepal_length", y= "sepal_width", data= phool)
plt.xlim(2)
plt.ylim(1)
plt.show()
```

### Styles
- Darkgrid
- Whitegrid
- Dark
- White
- Ticks

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt

# load dataset
phool = sns.load_dataset("iris")
phool

#make a line plot
sns.lineplot(x ="sepal_length", y= "sepal_width", data= phool)
#sns.set_theme() #use this code if we want to change in set style
sns.set_style("dark")
sns.set_style(style = None, rc= None)
plt.show()
```

### Size of figure

```python
# import libraries
import seaborn as sns
import matplotlib.pyplot as plt
# sns.set_style(style = None, rc= None)
# load dataset
phool = sns.load_dataset("iris")
phool
#change figure size
plt.figure(figsize=(6,4))

#make a line plot
sns.lineplot(x ="sepal_length", y= "sepal_width", data= phool)
plt.show()
```
